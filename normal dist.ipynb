```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
```

```python
# 3. Randomly sample 50 Starbucks customers. Let
# X = number of customers preferring lattes to cappuccinos
# Characteristics in common:
#  Each random variable counts number of successes in fixed number of trials.
#  Each trial has equal probability of success.
#  The outcome of each trial is independent.

#!/usr/bin/python

mu = 0
sigma = 1

class RandEvent:
    'my random event class'
    
    def __init__(self, myseed, mu, sigma, mycount):
        self.myrand = np.random.normal(mu, sigma, mycount)
        self.df = pd.DataFrame(self.myrand)
        self.myhist = self.df.hist()


def seedLoop():
#     print('seedloop')

    seeds = [1988,10,4343,3355, 8754]
    for seed in seeds:
        new_rand = RandEvent(seed, mu, sigma, 1000)
#         print('hello')
        printHist(new_rand)
    
def printHist(new_rand):
    df = pd.DataFrame(new_rand.myrand)
    df.hist()
    plt.show()
        
def main():
    print('main')
    seedLoop()

  
if __name__ == '__main__':
    main()
        
        
#         >> mu, sigma = 0, 0.1 # mean and standard deviation
# >>> s = np.random.normal(mu, sigma, 1000)

```
